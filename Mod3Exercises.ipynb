### Exercise 7:

Using the dataframe from the previous exercise, pick a method (Standard Deviation or Interquartile Range) and remove the outliers from the "FuelEconomy" column.

```python
# Fix FuelEconomy column to get rid of nulls and '-'. Make FuelEconomy numeric.
#Standard Deviation Method

df=df[df['FuelEconomy'].notnull()]
df = df[~df['FuelEconomy'].isin(['-'])]
df['FuelEconomy'] = pd.to_numeric(df['FuelEconomy'])

meangrade = df['FuelEconomy'].mean()
stdgrade = df['FuelEconomy'].std()
toprange = meangrade + stdgrade * 1.96
botrange = meangrade - stdgrade * 1.96

copydf = df.copy() #to not mess up the original df
copydf = copydf.drop(copydf[copydf['FuelEconomy'] > toprange].index)
copydf = copydf.drop(copydf[copydf['FuelEconomy'] < botrange].index)

copydf.head()
```

```python
#Interquartile Range Method

q1 = df['FuelEconomy'].quantile(.25)
q3 = df['FuelEconomy'].quantile(.75)
iqr = q3-q1
toprange = q3 + iqr * 1.5
botrange = q1 - iqr * 1.5

copydf = df.copy()
copydf = copydf.drop(copydf[copydf['FuelEconomy'] > toprange].index)
copydf = copydf.drop(copydf[copydf['FuelEconomy'] < botrange].index)

copydf.head()
```
